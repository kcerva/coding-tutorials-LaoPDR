# Data visualisation

In this notebook we describe some examples of plotting data using python. [This](https://towardsdatascience.com/introduction-to-data-visualization-in-python-89a54c97fbed0) is a good source to learn more about python packages for data visualistion.


## Using `pandas` (plot)

```python
#Import relevant packages for data analysis
import pandas as pd
import numpy as np

#import visualisation package
import matplotlib as plt
%matplotlib inline
```

```python
#simple plot 
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
#another dataframe visualisation
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
df.plot()
```

```python
#Plot data as a bar chart
df2 = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])

df2.plot(kind="bar")
```

```python
#Plot the same dataset as a horizontal bar chart
df2.plot(kind="barh")
```
